```python
# LCapy Netlist generated from PySpice
from lcapy import Circuit
cct = Circuit()

cct.add('Vdd 1 g1; down')
cct.add('M1 1 2 3; up')
cct.add('Rload 3 g2; right')

# Draw the circuit
cct.draw()
```
